```python
import pandas as pd
import plotly.graph_objects as go
import dash
from dash import dcc
from dash import html
from dash.dependencies import Input, Output
import plotly.express as px

import requests
import seaborn as sns
import pandas_ta as ta
```

```python
coin_pair = "btcusd"
url = f"https://www.bitstamp.net/api/v2/ohlc/{coin_pair}/"

params = {
        "step":60,
        "limit":int("30") + 28,
        }

data = requests.get(url, params=params).json()["data"]["ohlc"]

data = pd.DataFrame(data)
data.timestamp = pd.to_datetime(data.timestamp, unit = "s")

data[["open", "high", "low", "close", "volume"]] = data[["open","high","low","close","volume"]].astype(float)
```

```python
data
```

```python
data = data.groupby('timestamp').agg({
    'open': 'first',
    'high': 'max',
    'low': 'min',
    'close': 'last',
    'volume': 'sum'
}).reset_index()
```

```python
data["RSI"] = ta.rsi(data["close"])

data = data.iloc[14:]
# data = data.iloc[range_values[0]:range_values[1]]
```

```python
data = data.groupby('timestamp').agg({
    'open': 'first',    # Taking the first 'open' value
    'high': 'max',      # Maximum of 'high'
    'low': 'min',       # Minimum of 'low'
    'close': 'last',    # Taking the last 'close' value
    'volume': 'sum'     # Summing the volume
}).reset_index()
```

```python
data.shape
```

```python
candles = go.Figure(data=[go.Candlestick(
    x=data['timestamp'],
    open=data['open'],
    high=data['high'],
    low=data['low'],
    close=data['close'],
    increasing_line_color='green',
    decreasing_line_color='red'
)])

# Adjust layout for better spacing
candles.update_layout(
    xaxis_rangeslider_visible=False,   # Remove range slider
    xaxis_type='category',             # Use category axis for non-overlapping timestamp
    width=1000,                        # Adjust the overall width
    height=500,                        # Adjust height
    xaxis=dict(tickmode='auto', nticks=20), # Adjust the number of ticks to reduce overlap
    template= "plotly_dark"
)

candles.update_layout(transition_duration = 500)

# indicator = px.line(x=data.timestamp, y=data.rsi, height = 300, template = "plotly_dark")
# indicator.update_layout(transition_duration = 500)

candles.show()
```
